```python
import pandas as pd

# Load the dataset
file_path = "/content/smart_traffic_management_dataset.csv"
df = pd.read_csv(file_path)

# Show basic info and the first few rows
df.info(), df.head()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
df['waiting_time'] = df.apply(
    lambda row: row['traffic_volume'] if row['signal_status'] == 'Red' else 0, axis=1
)
```

```python
volume_max = df['traffic_volume'].max()
df['congestion_score'] = df['traffic_volume'] / volume_max
```

```python
grouped = df.groupby('location_id').agg({
    'waiting_time': 'mean',
    'avg_vehicle_speed': 'mean',
    'congestion_score': 'mean'
}).reset_index()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
normalized = scaler.fit_transform(grouped[['waiting_time', 'avg_vehicle_speed', 'congestion_score']])
grouped[['norm_waiting_time', 'norm_avg_speed', 'norm_congestion']] = normalized
```

```python
# Using equal weights: α = β = γ = 1/3
alpha = beta = gamma = 1/3

# Compute fitness for each location
grouped['fitness'] = (
    alpha * grouped['norm_waiting_time'] +
    beta * (1 - grouped['norm_avg_speed']) +  # higher speed = better, so subtract from 1
    gamma * grouped['norm_congestion']
)

# Sort by fitness to identify better-performing locations (lower is better)
grouped_sorted = grouped.sort_values(by='fitness').reset_index(drop=True)

print(grouped_sorted)  # or just
grouped_sorted.head()  # to view the top rows
```

```python
import numpy as np

# Parameters for GA
num_locations = grouped.shape[0]  # 5 intersections
population_size = 10  # You can increase this later

# Function to initialize population
def initialize_population(size, num_genes):
    """
    Randomly initialize a population of chromosomes.
    Each chromosome is a list of green time ratios between 0.1 and 0.9.
    """
    return np.random.uniform(low=0.1, high=0.9, size=(size, num_genes))

# Initialize the population
population = initialize_population(population_size, num_locations)
population
```

```python
# Extract base normalized metrics from grouped DataFrame
base_waiting = grouped['norm_waiting_time'].values
base_speed = grouped['norm_avg_speed'].values
base_congestion = grouped['norm_congestion'].values

# Function to evaluate fitness of a single chromosome
def evaluate_fitness(chromosome, alpha=1/3, beta=1/3, gamma=1/3):
    """
    Evaluate the fitness of a traffic signal plan (chromosome).
    More green time => reduces waiting & congestion, increases speed.
    """
    # Adjust metrics based on green time (simulated effect)
    adj_waiting = base_waiting * (1 - chromosome)
    adj_speed = base_speed * chromosome
    adj_congestion = base_congestion * (1 - chromosome)

    # Normalize speed again (higher speed = better, so use 1 - speed)
    fitness = alpha * np.mean(adj_waiting) + beta * np.mean(1 - adj_speed) + gamma * np.mean(adj_congestion)
    return fitness

# Evaluate fitness for all individuals in the population
fitness_scores = np.array([evaluate_fitness(ind) for ind in population])
fitness_scores
```

```python
# --- SELECTION ---
def select_parents(population, fitness_scores, num_parents=4):
    """
    Select the top 'num_parents' individuals based on fitness (lower is better).
    """
    # Get indices of best fitness scores
    best_indices = np.argsort(fitness_scores)[:num_parents]
    return population[best_indices]

# --- CROSSOVER ---
def crossover(parents, offspring_size):
    """
    Perform single-point crossover between pairs of parents.
    offspring_size = (number of offspring, number of genes)
    """
    offspring = np.empty(offspring_size)
    num_parents = parents.shape[0]
    num_genes = parents.shape[1]

    for k in range(offspring_size[0]):
        # Pick two parents randomly
        parent1_idx = k % num_parents
        parent2_idx = (k + 1) % num_parents
        crossover_point = np.random.randint(1, num_genes - 1)

        # Combine genes from both parents
        offspring[k, 0:crossover_point] = parents[parent1_idx, 0:crossover_point]
        offspring[k, crossover_point:] = parents[parent2_idx, crossover_point:]

    return offspring

# Select top parents
num_parents = 4
parents = select_parents(population, fitness_scores, num_parents=num_parents)

# Generate offspring via crossover
num_offspring = population_size - num_parents
offspring = crossover(parents, (num_offspring, num_locations))

parents, offspring
```

```python
 # --- MUTATION ---
def mutate(offspring, mutation_rate=0.2):
    """
    Randomly mutate genes in the offspring.
    Each gene has a chance to be slightly adjusted within a safe range.
    """
    num_offspring, num_genes = offspring.shape

    for i in range(num_offspring):
        for j in range(num_genes):
            if np.random.rand() < mutation_rate:
                # Apply a small random change
                mutation_value = np.random.uniform(-0.1, 0.1)
                offspring[i, j] += mutation_value
                # Keep values within bounds (0.1 to 0.9)
                offspring[i, j] = np.clip(offspring[i, j], 0.1, 0.9)

    return offspring

# Apply mutation to offspring
mutated_offspring = mutate(offspring.copy())

mutated_offspring
```

```python
import matplotlib.pyplot as plt

# GA parameters
generations = 20
population_size = 10
num_parents = 4
mutation_rate = 0.2

# Initialize the population
population = initialize_population(population_size, num_locations)
best_fitness_per_generation = []

for generation in range(generations):
    # Evaluate fitness
    fitness_scores = np.array([evaluate_fitness(ind) for ind in population])

    # Record best fitness in this generation
    best_fitness = np.min(fitness_scores)
    best_fitness_per_generation.append(best_fitness)

    # Select parents
    parents = select_parents(population, fitness_scores, num_parents=num_parents)

    # Crossover to create offspring
    num_offspring = population_size - num_parents
    offspring = crossover(parents, (num_offspring, num_locations))

    # Mutation
    mutated_offspring = mutate(offspring.copy(), mutation_rate)

    # New population = parents + mutated offspring
    population = np.vstack((parents, mutated_offspring))

# Final evaluation
final_fitness_scores = np.array([evaluate_fitness(ind) for ind in population])
best_solution = population[np.argmin(final_fitness_scores)]

# Plotting best fitness over generations
plt.figure(figsize=(10, 6))
plt.plot(best_fitness_per_generation, marker='o')
plt.title('Best Fitness Over Generations')
plt.xlabel('Generation')
plt.ylabel('Fitness (lower is better)')
plt.grid(True)
plt.tight_layout()
plt.show()

best_solution, min(best_fitness_per_generation)
```

```python
# PSO Parameters
num_particles = 10
dimensions = num_locations
iterations = 20
w = 0.5       # Inertia weight
c1 = 1.5      # Cognitive coefficient
c2 = 1.5      # Social coefficient

# Initialize particles
positions = initialize_population(num_particles, dimensions)  # Initial positions
velocities = np.random.uniform(-0.1, 0.1, size=(num_particles, dimensions))  # Initial velocities

# Initialize personal and global bests
personal_best_positions = positions.copy()
personal_best_scores = np.array([evaluate_fitness(p) for p in positions])
global_best_position = personal_best_positions[np.argmin(personal_best_scores)]
global_best_score = np.min(personal_best_scores)

# Track best fitness over time
pso_best_fitness = [global_best_score]

# PSO loop
for iter in range(iterations):
    for i in range(num_particles):
        # Update velocity
        r1, r2 = np.random.rand(), np.random.rand()
        velocities[i] = (
            w * velocities[i] +
            c1 * r1 * (personal_best_positions[i] - positions[i]) +
            c2 * r2 * (global_best_position - positions[i])
        )

        # Update position
        positions[i] += velocities[i]
        positions[i] = np.clip(positions[i], 0.1, 0.9)

        # Evaluate new fitness
        fitness = evaluate_fitness(positions[i])
        if fitness < personal_best_scores[i]:
            personal_best_positions[i] = positions[i].copy()
            personal_best_scores[i] = fitness

            if fitness < global_best_score:
                global_best_position = positions[i].copy()
                global_best_score = fitness

    pso_best_fitness.append(global_best_score)

# Plot PSO performance
plt.figure(figsize=(10, 6))
plt.plot(pso_best_fitness, marker='s', color='green')
plt.title('PSO: Best Fitness Over Generations')
plt.xlabel('Iteration')
plt.ylabel('Fitness (lower is better)')
plt.grid(True)
plt.tight_layout()
plt.show()

global_best_position, global_best_score
```

```python
# SA Parameters
initial_temp = 1.0
final_temp = 0.001
alpha = 0.9  # Cooling rate
max_iter = 100

# Generate an initial random solution (green time ratios)
current_solution = np.random.uniform(0.1, 0.9, size=num_locations)
current_fitness = evaluate_fitness(current_solution)
best_solution_sa = current_solution.copy()
best_fitness_sa = current_fitness

# Track fitness over iterations
sa_fitness_history = [current_fitness]

temp = initial_temp
iteration = 0

while temp > final_temp and iteration < max_iter:
    # Generate a neighbor by small random tweak
    neighbor = current_solution + np.random.uniform(-0.05, 0.05, size=num_locations)
    neighbor = np.clip(neighbor, 0.1, 0.9)
    neighbor_fitness = evaluate_fitness(neighbor)

    # Acceptance probability
    delta = neighbor_fitness - current_fitness
    if delta < 0 or np.random.rand() < np.exp(-delta / temp):
        current_solution = neighbor
        current_fitness = neighbor_fitness

        # Update best if improved
        if current_fitness < best_fitness_sa:
            best_solution_sa = current_solution.copy()
            best_fitness_sa = current_fitness

    sa_fitness_history.append(best_fitness_sa)
    temp *= alpha
    iteration += 1

# Plot SA performance
plt.figure(figsize=(10, 6))
plt.plot(sa_fitness_history, marker='^', color='orange')
plt.title('Simulated Annealing: Best Fitness Over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Fitness (lower is better)')
plt.grid(True)
plt.tight_layout()
plt.show()

best_solution_sa, best_fitness_sa
```

```python
# ACO Parameters
num_ants = 10
iterations = 20
alpha_pheromone = 1.0  # Influence of pheromone
beta_heuristic = 2.0   # Influence of heuristic (fitness)
evaporation_rate = 0.3
num_bins = 10  # Discrete green time options per location

# Create discrete green time values and initialize pheromones
green_time_values = np.linspace(0.1, 0.9, num_bins)
pheromone_matrix = np.ones((num_locations, num_bins))  # 5 intersections x 10 options

best_fitness_aco = float('inf')
best_solution_aco = None
aco_fitness_history = []

def construct_solution(pheromone_matrix):
    solution = []
    for i in range(num_locations):
        pheromones = pheromone_matrix[i] ** alpha_pheromone
        desirability = 1 / (np.arange(1, num_bins + 1)) ** beta_heuristic  # Heuristic just for variation
        prob = pheromones * desirability
        prob /= prob.sum()
        index = np.random.choice(range(num_bins), p=prob)
        solution.append(green_time_values[index])
    return np.array(solution)

for iteration in range(iterations):
    all_solutions = []
    all_fitnesses = []

    for _ in range(num_ants):
        solution = construct_solution(pheromone_matrix)
        fitness = evaluate_fitness(solution)
        all_solutions.append(solution)
        all_fitnesses.append(fitness)

        # Update best
        if fitness < best_fitness_aco:
            best_fitness_aco = fitness
            best_solution_aco = solution

    # Pheromone evaporation
    pheromone_matrix *= (1 - evaporation_rate)

    # Pheromone update
    for solution, fitness in zip(all_solutions, all_fitnesses):
        for i, green_time in enumerate(solution):
            bin_index = np.argmin(np.abs(green_time_values - green_time))
            pheromone_matrix[i][bin_index] += 1.0 / fitness  # Reinforce based on quality

    aco_fitness_history.append(best_fitness_aco)

# Plot ACO performance
plt.figure(figsize=(10, 6))
plt.plot(aco_fitness_history, marker='d', color='purple')
plt.title('ACO: Best Fitness Over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Fitness (lower is better)')
plt.grid(True)
plt.tight_layout()
plt.show()

best_solution_aco, best_fitness_aco
```

```python
# PSO Parameter Grid
pso_particle_sizes = [10, 20]
pso_iterations_list = [10, 20]
pso_inertia_weights = [0.4, 0.7]

# Store results
pso_tuning_results = []

# Run grid search for PSO
for particle_size in pso_particle_sizes:
    for iterations in pso_iterations_list:
        for inertia_weight in pso_inertia_weights:
            # Initialize positions and velocities
            positions = initialize_population(particle_size, num_locations)
            velocities = np.random.uniform(-0.1, 0.1, size=(particle_size, num_locations))

            personal_best_positions = positions.copy()
            personal_best_scores = np.array([evaluate_fitness(p) for p in positions])
            global_best_position = personal_best_positions[np.argmin(personal_best_scores)]
            global_best_score = np.min(personal_best_scores)

            for iter in range(iterations):
                for i in range(particle_size):
                    r1, r2 = np.random.rand(), np.random.rand()
                    velocities[i] = (
                        inertia_weight * velocities[i] +
                        1.5 * r1 * (personal_best_positions[i] - positions[i]) +
                        1.5 * r2 * (global_best_position - positions[i])
                    )
                    positions[i] += velocities[i]
                    positions[i] = np.clip(positions[i], 0.1, 0.9)

                    fitness = evaluate_fitness(positions[i])
                    if fitness < personal_best_scores[i]:
                        personal_best_positions[i] = positions[i].copy()
                        personal_best_scores[i] = fitness

                        if fitness < global_best_score:
                            global_best_position = positions[i].copy()
                            global_best_score = fitness

            # Store result
            pso_tuning_results.append({
                'Particles': particle_size,
                'Iterations': iterations,
                'Inertia Weight': inertia_weight,
                'Best Fitness': global_best_score
            })

# Convert to DataFrame for display
pso_tuning_df = pd.DataFrame(pso_tuning_results).sort_values(by='Best Fitness')
pso_tuning_df.head(10)
```

```python
# Hybrid GA + PSO parameters
ga_population = 10
ga_generations = 10
ga_mutation_rate = 0.3
pso_particles = 4
pso_iterations = 10
pso_inertia = 0.4

# Initialize GA population
population = initialize_population(ga_population, num_locations)
hybrid_fitness_history = []

for generation in range(ga_generations):
    # --- GA Phase ---
    fitness_scores = np.array([evaluate_fitness(ind) for ind in population])
    best_fitness = np.min(fitness_scores)
    hybrid_fitness_history.append(best_fitness)

    # Select top parents
    num_parents = int(ga_population / 2)
    parents = select_parents(population, fitness_scores, num_parents)

    # Crossover and Mutation
    offspring = crossover(parents, (ga_population - num_parents, num_locations))
    mutated_offspring = mutate(offspring.copy(), ga_mutation_rate)

    # New GA population
    population = np.vstack((parents, mutated_offspring))

    # --- PSO Phase ---
    # Take top-k from GA population and treat as PSO particles
    top_indices = np.argsort(np.array([evaluate_fitness(ind) for ind in population]))[:pso_particles]
    pso_positions = population[top_indices].copy()
    pso_velocities = np.random.uniform(-0.1, 0.1, size=(pso_particles, num_locations))

    pso_personal_best = pso_positions.copy()
    pso_personal_scores = np.array([evaluate_fitness(p) for p in pso_positions])
    pso_global_best = pso_personal_best[np.argmin(pso_personal_scores)]
    pso_global_score = np.min(pso_personal_scores)

    for _ in range(pso_iterations):
        for i in range(pso_particles):
            r1, r2 = np.random.rand(), np.random.rand()
            pso_velocities[i] = (
                pso_inertia * pso_velocities[i] +
                1.5 * r1 * (pso_personal_best[i] - pso_positions[i]) +
                1.5 * r2 * (pso_global_best - pso_positions[i])
            )
            pso_positions[i] += pso_velocities[i]
            pso_positions[i] = np.clip(pso_positions[i], 0.1, 0.9)

            fitness = evaluate_fitness(pso_positions[i])
            if fitness < pso_personal_scores[i]:
                pso_personal_best[i] = pso_positions[i].copy()
                pso_personal_scores[i] = fitness
                if fitness < pso_global_score:
                    pso_global_best = pso_positions[i].copy()
                    pso_global_score = fitness

    # Replace worst GA individuals with PSO-refined ones
    worst_indices = np.argsort(fitness_scores)[-pso_particles:]
    for i, idx in enumerate(worst_indices):
        population[idx] = pso_personal_best[i]

# Final result
final_fitness_scores = np.array([evaluate_fitness(ind) for ind in population])
best_hybrid_solution = population[np.argmin(final_fitness_scores)]
best_hybrid_fitness = np.min(final_fitness_scores)

# Plot fitness trend
plt.figure(figsize=(10, 6))
plt.plot(hybrid_fitness_history, marker='*', color='darkcyan')
plt.title("Hybrid GA + PSO: Fitness Over Generations")
plt.xlabel("Generation")
plt.ylabel("Fitness (lower is better)")
plt.grid(True)
plt.tight_layout()
plt.show()

best_hybrid_solution, best_hybrid_fitness
```

```python
# Hybrid GA + PSO for Traffic Signal Optimization
import numpy as np

# ---------------------- CONFIGURATION ----------------------
num_locations = 5  # Number of intersections
GA_POPULATION = 10
GA_GENERATIONS = 10
GA_MUTATION_RATE = 0.3
PSO_PARTICLES = 4
PSO_ITERATIONS = 10
PSO_INERTIA = 0.4

# ---------------------- FITNESS FUNCTION ----------------------
def evaluate_fitness(chromosome):
    base_waiting = np.array([0.44, 1.0, 0.0, 0.23, 0.35])
    base_speed = np.array([0.95, 0.71, 1.0, 0.0, 0.17])
    base_congestion = np.array([0.98, 1.0, 0.0, 0.72, 0.52])

    adj_waiting = base_waiting * (1 - chromosome)
    adj_speed = base_speed * chromosome
    adj_congestion = base_congestion * (1 - chromosome)

    fitness = (1/3) * np.mean(adj_waiting) + (1/3) * np.mean(1 - adj_speed) + (1/3) * np.mean(adj_congestion)
    return fitness

# ---------------------- GA HELPERS ----------------------
def initialize_population(size, genes):
    return np.random.uniform(0.1, 0.9, (size, genes))

def select_parents(pop, fitness, num_parents):
    best_idx = np.argsort(fitness)[:num_parents]
    return pop[best_idx]

def crossover(parents, offspring_size):
    offspring = np.empty(offspring_size)
    for k in range(offspring_size[0]):
        p1 = k % parents.shape[0]
        p2 = (k + 1) % parents.shape[0]
        cp = np.random.randint(1, offspring_size[1] - 1)
        offspring[k, :cp] = parents[p1, :cp]
        offspring[k, cp:] = parents[p2, cp:]
    return offspring

def mutate(offspring, rate):
    for i in range(offspring.shape[0]):
        for j in range(offspring.shape[1]):
            if np.random.rand() < rate:
                mutation = np.random.uniform(-0.1, 0.1)
                offspring[i, j] = np.clip(offspring[i, j] + mutation, 0.1, 0.9)
    return offspring

# ---------------------- HYBRID GA + PSO ----------------------
population = initialize_population(GA_POPULATION, num_locations)
best_fitness_history = []

for generation in range(GA_GENERATIONS):
    fitness_scores = np.array([evaluate_fitness(ind) for ind in population])
    best_fitness = np.min(fitness_scores)
    best_fitness_history.append(best_fitness)

    parents = select_parents(population, fitness_scores, GA_POPULATION // 2)
    offspring = crossover(parents, (GA_POPULATION - parents.shape[0], num_locations))
    mutated = mutate(offspring, GA_MUTATION_RATE)
    population = np.vstack((parents, mutated))

    # PSO Refinement
    top_idx = np.argsort([evaluate_fitness(ind) for ind in population])[:PSO_PARTICLES]
    positions = population[top_idx].copy()
    velocities = np.random.uniform(-0.1, 0.1, (PSO_PARTICLES, num_locations))

    pbest = positions.copy()
    pbest_scores = np.array([evaluate_fitness(p) for p in positions])
    gbest = pbest[np.argmin(pbest_scores)]
    gbest_score = np.min(pbest_scores)

    for _ in range(PSO_ITERATIONS):
        for i in range(PSO_PARTICLES):
            r1, r2 = np.random.rand(), np.random.rand()
            velocities[i] = (
                PSO_INERTIA * velocities[i] +
                1.5 * r1 * (pbest[i] - positions[i]) +
                1.5 * r2 * (gbest - positions[i])
            )
            positions[i] += velocities[i]
            positions[i] = np.clip(positions[i], 0.1, 0.9)

            fitness = evaluate_fitness(positions[i])
            if fitness < pbest_scores[i]:
                pbest[i] = positions[i].copy()
                pbest_scores[i] = fitness
                if fitness < gbest_score:
                    gbest = positions[i].copy()
                    gbest_score = fitness

    worst_idx = np.argsort(fitness_scores)[-PSO_PARTICLES:]
    for i, idx in enumerate(worst_idx):
        population[idx] = pbest[i]

# ---------------------- FINAL RESULT ----------------------
final_fitness_scores = np.array([evaluate_fitness(ind) for ind in population])
best_solution = population[np.argmin(final_fitness_scores)]
best_fitness = np.min(final_fitness_scores)

print("Best Signal Plan (Green Time Ratios):", best_solution)
print("Best Fitness Score:", best_fitness)
```

```python
# GA + SA Hybrid Parameters
ga_population = 10
ga_generations = 15
mutation_rate = 0.3

sa_top_k = 2  # Number of top individuals to refine with SA
sa_initial_temp = 1.0
sa_final_temp = 0.001
sa_alpha = 0.9
sa_max_steps = 30

# Initialize GA population
population = initialize_population(ga_population, num_locations)
ga_sa_fitness_history = []

for generation in range(ga_generations):
    # Evaluate fitness
    fitness_scores = np.array([evaluate_fitness(ind) for ind in population])
    best_fitness = np.min(fitness_scores)
    ga_sa_fitness_history.append(best_fitness)

    # Select top parents
    num_parents = int(ga_population / 2)
    parents = select_parents(population, fitness_scores, num_parents)

    # Crossover and mutation
    offspring = crossover(parents, (ga_population - num_parents, num_locations))
    mutated_offspring = mutate(offspring.copy(), mutation_rate)

    # Combine to form new population
    population = np.vstack((parents, mutated_offspring))

    # --- Simulated Annealing Phase on top-k individuals ---
    top_indices = np.argsort(np.array([evaluate_fitness(ind) for ind in population]))[:sa_top_k]
    for idx in top_indices:
        current = population[idx].copy()
        current_fitness = evaluate_fitness(current)
        best_sa = current.copy()
        best_sa_fitness = current_fitness
        temp = sa_initial_temp

        while temp > sa_final_temp:
            for _ in range(sa_max_steps):
                neighbor = current + np.random.uniform(-0.05, 0.05, size=num_locations)
                neighbor = np.clip(neighbor, 0.1, 0.9)
                neighbor_fitness = evaluate_fitness(neighbor)

                delta = neighbor_fitness - current_fitness
                if delta < 0 or np.random.rand() < np.exp(-delta / temp):
                    current = neighbor.copy()
                    current_fitness = neighbor_fitness

                    if current_fitness < best_sa_fitness:
                        best_sa = current.copy()
                        best_sa_fitness = current_fitness

            temp *= sa_alpha

        # Replace in population
        population[idx] = best_sa

# Final evaluation
final_fitness_scores = np.array([evaluate_fitness(ind) for ind in population])
best_ga_sa_solution = population[np.argmin(final_fitness_scores)]
best_ga_sa_fitness = np.min(final_fitness_scores)

# Plot GA + SA hybrid performance
plt.figure(figsize=(10, 6))
plt.plot(ga_sa_fitness_history, marker='P', color='darkorange')
plt.title("Hybrid GA + SA: Fitness Over Generations")
plt.xlabel("Generation")
plt.ylabel("Fitness (lower is better)")
plt.grid(True)
plt.tight_layout()
plt.show()

best_ga_sa_solution, best_ga_sa_fitness
```

```python
# Compare both hybrid models: GA+PSO vs. GA+SA
# Reuse GA+SA history (ga_sa_fitness_history), and regenerate GA+PSO history with same setup

# GA + PSO parameters (same as before)
ga_population = 10
ga_generations = 15
mutation_rate = 0.3

pso_particles = 4
pso_iterations = 10
pso_inertia = 0.4

# Reinitialize population
population = initialize_population(ga_population, num_locations)
ga_pso_fitness_history = []

for generation in range(ga_generations):
    fitness_scores = np.array([evaluate_fitness(ind) for ind in population])
    best_fitness = np.min(fitness_scores)
    ga_pso_fitness_history.append(best_fitness)

    num_parents = int(ga_population / 2)
    parents = select_parents(population, fitness_scores, num_parents)
    offspring = crossover(parents, (ga_population - num_parents, num_locations))
    mutated_offspring = mutate(offspring.copy(), mutation_rate)
    population = np.vstack((parents, mutated_offspring))

    # PSO refinement on top-k
    top_indices = np.argsort(np.array([evaluate_fitness(ind) for ind in population]))[:pso_particles]
    pso_positions = population[top_indices].copy()
    pso_velocities = np.random.uniform(-0.1, 0.1, size=(pso_particles, num_locations))

    pso_personal_best = pso_positions.copy()
    pso_personal_scores = np.array([evaluate_fitness(p) for p in pso_positions])
    pso_global_best = pso_personal_best[np.argmin(pso_personal_scores)]
    pso_global_score = np.min(pso_personal_scores)

    for _ in range(pso_iterations):
        for i in range(pso_particles):
            r1, r2 = np.random.rand(), np.random.rand()
            pso_velocities[i] = (
                pso_inertia * pso_velocities[i] +
                1.5 * r1 * (pso_personal_best[i] - pso_positions[i]) +
                1.5 * r2 * (pso_global_best - pso_positions[i])
            )
            pso_positions[i] += pso_velocities[i]
            pso_positions[i] = np.clip(pso_positions[i], 0.1, 0.9)

            fitness = evaluate_fitness(pso_positions[i])
            if fitness < pso_personal_scores[i]:
                pso_personal_best[i] = pso_positions[i].copy()
                pso_personal_scores[i] = fitness
                if fitness < pso_global_score:
                    pso_global_best = pso_positions[i].copy()
                    pso_global_score = fitness

    worst_indices = np.argsort(fitness_scores)[-pso_particles:]
    for i, idx in enumerate(worst_indices):
        population[idx] = pso_personal_best[i]

# Plot comparison
plt.figure(figsize=(10, 6))
plt.plot(ga_sa_fitness_history, label="Hybrid GA + SA", marker='P', color='orange')
plt.plot(ga_pso_fitness_history, label="Hybrid GA + PSO", marker='o', color='blue')
plt.title("Comparison: Hybrid GA + SA vs Hybrid GA + PSO")
plt.xlabel("Generation")
plt.ylabel("Fitness (lower is better)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
